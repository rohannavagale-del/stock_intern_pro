```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('Zomato data .csv')
df
```

```python
df.head()
```

convert the data type of column rate 

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handleRate)
df
```

```python
df.isnull().sum()
```

Type of restraunt

```python
sns.countplot(x = df['listed_in(type)'])
plt.xlabel("Type Of Restraunt")
plt.show()
```

conclusion - majority of the restraunt falls in dinning 

```python
df.head()
```

which restarunt the most votes 

```python
group_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : group_data})
plt.plot(result , c = 'green' , marker = '*')
plt.xlabel("Type of restraunt" , size = 20)
plt.ylabel("No of votes" , size = 20)
```

conclusion - Dining restraunt has recived max no of votes

which mode recive max rating 

```python
sns.boxplot( x = 'online_order' , y = 'rate' , data = df)
plt.show()
```

how many pepole order online or offline 

```python
pivot_table = df.pivot_table(index= 'listed_in(type)' , columns= 'online_order' , aggfunc='size' , fill_value=0)
sns.heatmap(pivot_table , annot= True , fmt='d')
plt.title("HeatMap")
plt.xlabel("online_order")
plt.ylabel("restraunt type") 
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

sns.countplot(x='listed_in(type)', data=df, ax=axes[0, 0])
sns.lineplot(x='votes', y='rate', data=df, ax=axes[0, 1])
sns.histplot(df['votes'], bins=30, ax=axes[0, 2])
sns.countplot(x='online_order', data=df, ax=axes[1, 0])
sns.boxplot(x='online_order', y='votes', data=df, ax=axes[1, 1])
sns.heatmap(pivot_table, annot=True, fmt='d', ax=axes[1, 2])


plt.tight_layout()
axes[0,0].set_title("Restaurant Type Count")
axes[0,1].set_title("Votes vs Rating")
axes[0,2].set_title("Votes Distribution")
axes[1,0].set_title("Online Order Count")
axes[1,1].set_title("Votes by Online Order")
axes[1,2].set_title("Type vs Online Order Heatmap")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("my_plot.png", bbox_inches="tight")

plt.show()



plt.show()

```
