Categorize Vote_Average column 
we would cut the Vote_Average values and make 4 categorie popular , Average , below_Average , not_popular to describe it more using categorize_col() function

```python
def Categorize_col(df , col , labels ): # create a function in which we are giving a lables to the col 
    edges= [df[col].describe()['min'],
            df[col].describe()['25%'],
            df[col].describe()['50%'],
            df[col].describe()['75%'],
            df[col].describe()['max']] 
    df[col] = pd.cut(df[col], edges , labels = labels )
    return df
```

```python
labels = ['not_popular' , 'below_avg' , 'average' , 'popular']# give labels to Vote_Average now in that col the labels will show
Categorize_col(df ,'Vote_Average', labels)
df['Vote_Average'].unique()
```

```python
sns.catplot(y = 'Genre' , data = df , kind = 'count',
            order = df['Genre'].value_counts().index ) # order shows the Genre values in order the most frequent one will come first
plt.title("Genre Column Distribution")
plt.xlabel("No of Counts")
plt.ylabel("Name of Genr")
plt.show()


```
